```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, Activation
from keras.layers import BatchNormalization as BN
from keras.callbacks import ModelCheckpoint, ReduceLROnPlateau
from keras.preprocessing.image import ImageDataGenerator
from keras.optimizers import Adam
from keras.applications import DenseNet201
from keras.metrics import TopKCategoricalAccuracy
```

```python
train_datagen = ImageDataGenerator(
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range = 0.2,
    horizontal_flip=True,
    validation_split=0.2)
```

```python
target_size = (64, 32)
batch_size = 512
mode = 'categorical'
numclasses = 291
```

```python
train_generator = train_datagen.flow_from_directory(
    data_train_dir,
    target_size=target_size,
    batch_size=batch_size,
    class_mode=mode,
    subset='training') # set as training data
```

```python
validation_generator = train_datagen.flow_from_directory(
    data_train_dir, # same directory as training data
    target_size=target_size,
    batch_size=batch_size,
    class_mode=mode,
    subset='validation') # set as validation data
```

```python
## DEF NN TOPOLOGY  
model = DenseNet201(weights=None, classes=291, input_shape=(*target_size, 3))
```

```python
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=10, min_lr=0.00001)
```

```python
checkpoint_cb = ModelCheckpoint(filepath="/content/drive/MyDrive/Colab Notebooks/insectos_denseNet201_adam/insectos_denseNet201_adam_weight", save_best_only=True, save_weights_only=True, monitor='val_accuracy')
```

```python
## OPTIM AND COMPILE

model.compile(loss='categorical_crossentropy',
              optimizer=Adam(),
              metrics=['accuracy', TopKCategoricalAccuracy(k=3, name="Top_3_accuracy")])
```

```python
## TRAINING with DA and LRA
history=model.fit(train_generator,
                            steps_per_epoch=len(train_generator.filenames) // batch_size, 
                            epochs=50,
                            validation_data=validation_generator,
                            callbacks=[reduce_lr,checkpoint_cb],
                            verbose=1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
